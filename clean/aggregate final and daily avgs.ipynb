```python
import pandas as pd
import numpy as np
```

```python
# read in raw novel word count data
df1 = pd.read_csv("../rawdata/novels_wc_info.csv", index_col=0)
df1.head()
```

```python
# read in the clean novel data
df2 = pd.read_csv("../clean data/novel_data.csv", index_col=0)
df2.head()
```

```python
df2['Novel Date'] = df1['Novel Date']

df2.head()
```

```python
# get a list of writer names
writers = pd.read_csv("../rawdata/user_profiles_basicinfo.csv")[['Writer Name']]
writers.head()
```

```python
# group the data by Writer Name
writergroups = df2.groupby('Writer Name')

expected_final_wc = []
expcted_daily_avg = []

# iterate through writer names
for i in writers['Writer Name']:
    # isolate a single writer and calculate their daily average
    group = writergroups.get_group(i)
    group['dailyavg'] = group['Final Word Count'] / 30.0
    
    #get the average of final word count and dailyavg and append the values 
    expected_final_wc.append(np.mean(group['Final Word Count'].values))
    expcted_daily_avg.append(np.mean(group['dailyavg']))

```

```python
expected_final_wc
```

```python
user_summary = pd.read_csv("../clean data/user_summary.csv", index_col=0)
user_summary.head()
```

```python
user_summary['Expected Final Word Count'] = expected_final_wc
user_summary['Expected Daily Average'] = expcted_daily_avg
user_summary
```

```python
user_summary.to_csv("../clean data/user_summary.csv")
```
