```python
import pandas as pd
```

```python
df = pd.read_csv("../rawdata/user_profiles_basicinfo.csv")
```

```python
df = df[['Writer Name', 'Member Length', 'LifetimeWordCount', 'url']]
df
```

```python
def getMemberLength(string):
    string = string.split("Member for ")[1]
    string = string.split(" ")
    if len(string) < 3:
        memberlength = int(string[0]) * (1.0/12.0)
    else:
        memberlength = int(string[1])
    return memberlength

getMemberLength(df['Member Length'][0])
```

```python
df['Member Length'] = df['Member Length'].apply(lambda x: getMemberLength(x))
df.head()
```

```python
facts = pd.read_csv("../clean data/fact_sheet_parsed.csv")
```

```python
facts = facts[['Age', 'Birthday', 'Favorite books or authors', 'Favorite noveling music', 'Hobbies', 'Location', 'Occupation', 'Primary Role', 'Sponsorship URL']]
```

```python
summary = pd.concat([df, facts], axis = 1)
```

```python
summary.to_csv('../clean data/user_summary.csv')
```
