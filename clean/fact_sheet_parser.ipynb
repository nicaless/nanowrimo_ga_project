```python
import pandas as pd
```

```python
# read in fact sheet raw data
df = pd.read_csv("../rawdata/user_profiles_factsheet.csv")
df.head()
```

```python
# get rid of the unnecessary rows
df = df[df['Fact Sheet'] != 'Fact Sheet']
```

```python
for facts in df['Fact Sheet'][0:40]:
    print facts
```

```python
allfacts = []

j = 0
for facts in df['Fact Sheet']:
    
    fact_sheet = {}
    
    # if fact sheet is empty if presents as an NaN float.  append it anyway and continue 
    if isinstance(facts, float):
        allfacts.append(fact_sheet)
        continue
    
    facts = facts.split(": ")
    
    # The 0th string in facts is the first fact_name
    fact_name = facts[0]
    
    for i in range(1, len(facts)):
        # string split fact_names from the fact information and append the facts to the fact_sheet dictionary
        if "Primary Role" in facts[i]:
            info = facts[i].split(" Primary Role")
            fact_sheet[fact_name] = ' '.join(info[:len(info)-1])
            fact_name = "Primary Role"
        elif "Sponsorship URL" in facts[i]:
            info = facts[i].split(" Sponsorship URL")
            fact_sheet[fact_name] = ' '.join(info[:len(info)-1])
            fact_name = "Sponsorship URL"
        elif "Favorite books or authors" in facts[i]:
            info = facts[i].split(" Favorite books or authors")
            fact_sheet[fact_name] = ' '.join(info[:len(info)-1])
            fact_name = "Favorite books or authors"
        elif "Favorite noveling music" in facts[i]:
            info = facts[i].split(" Favorite noveling music")
            fact_sheet[fact_name] = ' '.join(info[:len(info)-1])
            fact_name = "Favorite noveling music"
        else:
            info = facts[i].split(" ")
            fact_sheet[fact_name] = ' '.join(info[:len(info)-1])
            fact_name = info[-1]
            
    allfacts.append(fact_sheet)
    
print allfacts
```

```python
my_df = pd.DataFrame([allfacts[0]])
```

```python
my_df
```

```python
# create one large data frame with all the user facts
for i in range(1, len(allfacts)):
    temp = pd.DataFrame([allfacts[i]])
    my_df = pd.concat([my_df, temp])
```

```python
# rearranging the columns
my_df = my_df[['Age', 'Birthday', 'Favorite books or authors', 'Favorite noveling music', 'Hobbies', 'Location', 'Occupation', 'Primary Role','Sponsorship URL']]
my_df
```
