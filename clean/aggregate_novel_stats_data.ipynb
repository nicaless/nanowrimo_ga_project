```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../rawdata/novels_wc_info.csv")
del df['Unnamed: 0']
df.head()
```

```python
writers = pd.read_csv("../rawdata/user_profiles_basicinfo.csv")[['url']]
writers['url'] = writers['url'].apply(lambda x: x + str('/novels'))
writers['url'].head()
```

```python
writergroups = df.groupby('Writer Url')
```

```python
nicaless = writergroups.get_group('http://nanowrimo.org/participants/nicaless/novels')
```

```python
np.mean(nicaless['Num Submissions'].values)
print len(nicaless)
```

```python
num_novels = []
expected_num_submissions = []
expected_avg_submission = []
expected_min_submission = []
expected_min_day = []
expected_max_submission = []
expected_max_day = []
expected_std = []
expected_consec_sub = []

for i in writers['url']:
    
    group = writergroups.get_group(i)
    
    num_novels.append(len(group))
    expected_num_submissions.append(np.mean(group['Num Submissions'].values))
    expected_avg_submission.append(np.mean(group['Avg Submission'].values))
    expected_min_submission.append(np.mean(group['Min Submission'].values))
    expected_min_day.append(np.mean(group['Min Day'].values))
    expected_max_submission.append(np.mean(group['Max Submission'].values))
    expected_max_day.append(np.mean(group['Max Day'].values))
    expected_std.append(np.mean(group['Standard Deviation'].values))
    expected_consec_sub.append(np.mean(group['Consecutive Submissions']))

```

```python
user_summary = pd.read_csv("../clean data/user_summary.csv")
del user_summary['Unnamed: 0']
del user_summary['Num Novels']
del user_summary['Expected Num Submissions']
del user_summary['Expected Avg Submission']
del user_summary['Expected Min Submission']
del user_summary['Expected Min Day']
del user_summary['Expected Max Submission']
del user_summary['Expected Max Day']
del user_summary['Expected Std Submissions']
del user_summary['Expected Consec Subs']
user_summary.head()
```

```python
user_summary['Num Novels'] = num_novels
user_summary['Expected Num Submissions'] = expected_num_submissions
user_summary['Expected Avg Submission'] = expected_avg_submission
user_summary['Expected Min Submission'] = expected_min_submission
user_summary['Expected Min Day'] = expected_min_day
user_summary['Expected Max Submission'] = expected_max_submission
user_summary['Expected Max Day'] = expected_max_day
user_summary['Expected Std Submissions'] = expected_std
user_summary['Expected Consec Subs'] = expected_consec_sub

user_summary.head()
```

```python
user_summary.to_csv("../clean data/user_summary.csv")
```
