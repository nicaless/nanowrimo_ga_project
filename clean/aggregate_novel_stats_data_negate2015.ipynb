```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../rawdata/novels_wc_info.csv")
del df['Unnamed: 0']
df = df[df['Novel Date'] != 'November 2015']
df.head()

'http://nanowrimo.org/participants/nicaless/novels' not in df['Writer Url'].values
```

```python
writers = pd.read_csv("../rawdata/user_profiles_basicinfo.csv")[['url']]
writers['url'] = writers['url'].apply(lambda x: x + str('/novels'))
writers['url'].head()
```

```python
writergroups = df.groupby('Writer Url')
```

```python
nicaless = writergroups.get_group('http://nanowrimo.org/participants/nicaless/novels')
```

```python
np.mean(nicaless['Num Submissions'].values)
print len(nicaless)
```

```python
num_novels = []
expected_num_submissions = []
expected_avg_submission = []
expected_min_submission = []
expected_min_day = []
expected_max_submission = []
expected_max_day = []
expected_std = []

for i in writers['url']:
    if i not in df['Writer Url'].values:
        num_novels.append(float('NaN'))
        expected_num_submissions.append(float('NaN'))
        expected_avg_submission.append(float('NaN'))
        expected_min_submission.append(float('NaN'))
        expected_min_day.append(float('NaN'))
        expected_max_submission.append(float('NaN'))
        expected_max_day.append(float('NaN'))
        expected_std.append(float('NaN'))
        continue
    
    group = writergroups.get_group(i)
    
    num_novels.append(len(group))
    expected_num_submissions.append(np.mean(group['Num Submissions'].values))
    expected_avg_submission.append(np.mean(group['Avg Submission'].values))
    expected_min_submission.append(np.mean(group['Min Submission'].values))
    expected_min_day.append(np.mean(group['Min Day'].values))
    expected_max_submission.append(np.mean(group['Max Submission'].values))
    expected_max_day.append(np.mean(group['Max Day'].values))
    expected_std.append(np.mean(group['Standard Deviation'].values))

```

```python
user_summary_no2015 = pd.read_csv("../clean data/user_summary_no2015.csv")
del user_summary_no2015['Unnamed: 0']
user_summary_no2015.head()
```

```python
user_summary_no2015['Num Novels'] = num_novels
user_summary_no2015['Expected Num Submissions'] = expected_num_submissions
user_summary_no2015['Expected Avg Submission'] = expected_avg_submission
user_summary_no2015['Expected Min Submission'] = expected_min_submission
user_summary_no2015['Expected Min Day'] = expected_min_day
user_summary_no2015['Expected Max Submission'] = expected_max_submission
user_summary_no2015['Expected Max Day'] = expected_max_day
user_summary_no2015['Expected Std Submissions'] = expected_std

user_summary_no2015.head()
```

```python
user_summary_no2015.to_csv("../clean data/user_summary_no2015.csv")
```
