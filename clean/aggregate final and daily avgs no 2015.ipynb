```python
#follows the same workflow as 'aggregate final and daily avgs.ipynb'
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("../rawdata/novels_wc_info.csv", index_col=0)
df1.head()
```

```python
df2 = pd.read_csv("../clean data/novel_data.csv", index_col=0)
df2.head()
```

```python
df2['Novel Date'] = df1['Novel Date']
# exclude all novels written November 2015
df2 = df2[df2['Novel Date'] != 'November 2015']

df2.head()
```

```python
writers = pd.read_csv("../rawdata/user_profiles_basicinfo.csv")[['Writer Name']]
writers.head()
```

```python
writergroups = df2.groupby('Writer Name')

expected_final_wc = []
expcted_daily_avg = []

for i in writers['Writer Name']:
    # if writer cannot be found in the df2 data frame, it is because they have only written a novel in November 2015
    # mark their entries as NaN and continue
    if i not in df2['Writer Name'].values:
        expected_final_wc.append(float('NaN'))
        expcted_daily_avg.append(float('NaN'))
        continue
    
    group = writergroups.get_group(i)
    group['dailyavg'] = group['Final Word Count'] / 30.0
    
    expected_final_wc.append(np.mean(group['Final Word Count'].values))
    expcted_daily_avg.append(np.mean(group['dailyavg']))

```

```python
expected_final_wc
```

```python
user_summary_no2015 = pd.read_csv("../clean data/user_summary_no2015.csv")
del user_summary_no2015['Unnamed: 0']
user_summary_no2015.head()
```

```python
user_summary_no2015['Expected Final Word Count'] = expected_final_wc
user_summary_no2015['Expected Daily Average'] = expcted_daily_avg
user_summary_no2015
```

```python
user_summary_no2015.to_csv("../clean data/user_summary_no2015.csv")
```
