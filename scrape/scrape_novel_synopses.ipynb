```python
import urllib
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
novel_pages = pd.read_csv('../rawdata/novel_pages.csv')
novel_pages.head()
```

```python
synopses = []
for i in novel_pages['Novel Pages']:
    print i
    page = urllib.urlopen(i).read()
    soup = BeautifulSoup(page, 'html.parser')
    synopsis = soup.find(id='novel_synopsis').encode_contents()
    synopses.append(synopsis)
```

```python
print synopses
print len(synopses)
```

```python
novel_data = pd.read_csv("../rawdata/novel_meta_data.csv")
novel_data.head()
```

```python
novel_data['Synopses'] = synopses
```

```python
novel_data.head()
```

```python
novel_data = novel_data[['Writer Name', 'Novel Name', 'Genre', 'Final Word Count', 'Average Per Day', 'Winner', 'Synopses', 'url']]
```
