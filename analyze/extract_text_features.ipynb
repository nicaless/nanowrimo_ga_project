```python
import pandas as pd
```

```python
novels = pd.read_csv("../clean data/novel_data.csv", index_col=0)
novels.head()
```

```python
# clean novel name
novel_names = []
for n in novels['Novel Name']:
    novel_names.append(n.split("Novel: ")[1])

novels['Novel Name'] = novel_names
novels['Novel Name'].head()
```

```python
genre = []
for g in novels['Genre']:
    genre.append(g.split("Genre:")[1])

novels['Genre'] = genre
novels['Genre'].head()
```

```python
standard_genres = ['Adventure','Children\'s Fiction','Erotic','Fanfiction','Fantasy','Historical','Horror/Supernatural','Literary','Mainstream','Mystery','Personal','Religious/Spiritual','Romance','Satire/Humor','Science Fiction','Thriller/Suspense','Women\'s Fiction','Young Adult']
```

```python
test = novels['Synopses'][1:10]
test
```

```python
has_synopses = []
num_words = []
num_unique = []
num_paragraphs = []
num_sentences = []
has_genre = []
standard_genre = []

from bs4 import BeautifulSoup

#for novel in test:
i = 0
for novel in novels['Synopses']:
    soup = BeautifulSoup(novel, 'html.parser')
    syn = soup.find_all("p")
    if syn[0].renderContents() == "":
        has_synopses.append(0)
        num_paragraphs.append(0)
    else:
        has_synopses.append(1)
        num_paragraphs.append(len(syn))
    num_sent = 0
    wc = 0
    uniques = 0
    for i in syn:
        line =  i.renderContents()
        num_sent = num_sent + line.count(".")
        wc = wc + len(line.split(" "))
        uniqueWords = []
        for j in line:
            if not j in uniqueWords:
                uniqueWords.append(i)
        uniques = uniques + len(uniqueWords)
    
    num_sentences.append(num_sent)
    num_words.append(wc)
    num_unique.append(uniques)


```

```python
for genre in novels['Genre']:
    if genre == "":
        has_genre.append(0)
    else:
        has_genre.append(1)
    if genre in standard_genres:
        standard_genre.append(1)
    else:
        standard_genre.append(0)
```

```python
print len(has_synopses)
print len(num_words)
print len(num_unique)
print len(num_paragraphs)
print len(num_sentences)
print len(has_genre)
print len(standard_genre)
```

```python
novel_features = pd.DataFrame(novels['Winner'])
```

```python
novel_features['has_genre'] = has_genre
novel_features['standard_genre'] = standard_genre
novel_features['has_synopses'] = has_synopses
novel_features['num_words'] = num_words
novel_features['num_unique'] = num_unique
novel_features['num_paragraphs'] = num_paragraphs
novel_features['num_sentences'] = num_sentences
```
