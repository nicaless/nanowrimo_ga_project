```python
excerpts = []
num_paragraphs = []
num_words = []
num_uniques = []
num_sentences = []
for novel in novels['Excerpt']:
#for novel in test:
    mystr = ""
    soup = BeautifulSoup(novel, 'html.parser')
    syn = soup.find_all("p")
    for i in syn:
        mystr = mystr + i.renderContents() + " "
    excerpts.append(mystr)
    num_words.append(textstat.lexicon_count(mystr))
    if textstat.lexicon_count(mystr) == 0:
        num_sentences.append(0)
        num_paragraphs.append(0)
        num_uniques.append(0)
    else:
        num_sentences.append(textstat.sentence_count(mystr))
        num_paragraphs.append(len(syn))
        num_uniques.append(len(Counter(mystr.split(" "))))
```

```python
scores = []
has_excerpt = []
for i in excerpts:
    if i == " " or textstat.sentence_count(i) == 0:
        scores.append(0)
        has_excerpt.append(0)
    else:
        scores.append(textstat.flesch_reading_ease(i))
        has_excerpt.append(1)
```

```python
novel_features = pd.read_csv("novel_features.csv", index_col = 0)
novel_features.head()
```

```python
#num_paragraphs
#num_words
#num_uniques
#scores
#has_excerpt
#num_sentences
novel_features['has excerpt'] = has_excerpt
novel_features['num words excerpt'] = num_words
novel_features['num uniques excerpt'] = num_uniques
novel_features['num sentences excerpt'] = num_sentences
novel_features['paragraphs excerpt'] = num_paragraphs
novel_features['fk score excerpt'] = scores
```
